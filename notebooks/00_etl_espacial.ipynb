```python
import geopandas as gpd
import os

# 1. LOCALIZADOR DE RUTAS (El mismo de antes para que no falle)
if os.path.exists('data'):
    PATH_DATA = 'data'
elif os.path.exists(os.path.join('..', 'data')):
    PATH_DATA = os.path.join('..', 'data')
else:
    PATH_DATA = 'data'

# 2. DEFINIR RUTAS DE ENTRADA Y SALIDA
ruta_shapefile_gigante = os.path.join(PATH_DATA, 'SECC_CE_20230101.shp')
ruta_geojson_salida = os.path.join(PATH_DATA, 'mapa_sevilla.geojson')

try:
    print("⏳ 1/3 Cargando el mapa gigante de España (esto puede tardar unos segundos)...")
    # Cargamos el shapefile original
    mapa_espana = gpd.read_file(ruta_shapefile_gigante)
    
    print("🔍 2/3 Buscando los barrios de Sevilla Capital...")
    # En los archivos del INE, la columna de 10 dígitos se llama 'CUSEC'
    # Filtramos las que empiezan por 41091 (Código de Sevilla capital)
    mapa_sevilla = mapa_espana[mapa_espana['CUSEC'].astype(str).str.startswith('41091')].copy()
    
    print("💾 3/3 Guardando el nuevo archivo optimizado...")
    # Lo guardamos como GeoJSON, un formato de un solo archivo, ideal para GitHub
    mapa_sevilla.to_file(ruta_geojson_salida, driver='GeoJSON')
    
    print(f"\n✅ ¡MISIÓN CUMPLIDA!")
    print(f"🗺️ Secciones guardadas: {len(mapa_sevilla)}")
    print(f"📁 Nuevo archivo listo para GitHub: {ruta_geojson_salida}")

except FileNotFoundError:
    print(f"❌ No encuentro el archivo gigante en {ruta_shapefile_gigante}")
except Exception as e:
    print(f"❌ Ha ocurrido un error: {e}")
```
